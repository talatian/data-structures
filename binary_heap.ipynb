```python
import math

class BinaryHeap:
    def __init__(self, values=[]):
        self.values = values
    
    def insert(self, value):
        
        def bubble_up():
            index = len(self.values) - 1
            parent = math.floor((index-1)/2)
            while self.values[index] > self.values[parent] and index > 0:
                self.values[index], self.values[parent] = self.values[parent], self.values[index]
                index = parent
                parent = math.floor((index-1)/2)
                
        self.values.append(value)
        bubble_up()
        return self.values
    
    def extract_max(self):
        def sink_down():
            index = 0
            while True:
                current_value = self.values[index]
                left_child_index = 2*index+1
                right_child_index = 2*index+2
                swap = None
                if left_child_index < len(self.values):
                    left_child = self.values[left_child_index]
                    if left_child > current_value:
                        swap = left_child_index
                if right_child_index < len(self.values):
                    right_child = self.values[right_child_index]
                    if (right_child > current_value and not swap) or (right_child > left_child and swap):
                        swap = right_child_index
                if not swap:
                    break
                self.values[index], self.values[swap] = self.values[swap], self.values[index]
                index = swap
        if self.values:        
            extracted = self.values[0]
            last = self.values.pop()
        else:
            extracted = None
        if self.values:
            self.values[0] = last
            sink_down()
        return extracted
        

    def __str__(self): 
        def traverse(node_index, level=0):
            if node_index >= len(self.values):
                return ''
            right_node_index = 2*node_index+2
            ret = traverse(right_node_index, level+1)
            ret += level * 10 * ' '
            ret += str(self.values[node_index]) + '\n'
            left_node_index = 2*node_index+1
            ret += traverse(left_node_index, level+1)
            return ret
        return traverse(0)    
        
```

```python
bh = BinaryHeap()
bh.insert(10)
bh.insert(15)
bh.insert(20)
bh.insert(100)
bh.insert(1)
bh.insert(2)
bh.insert(22)


print(bh)

print(bh.extract_max())
print('\n'*3)
print(bh)
print(bh.extract_max())
print('\n'*3)
print(bh)
```
