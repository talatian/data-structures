```python
class Node:
    
    def __init__(self, value):
        self.value = value
        self.next = None
    
    def __str__(self):
        return f"value: {self.value}"
    
    def __repr__(self):
        return f"Node( {self.value})"


class Stack:
    
    def __init__(self):
        self.length = 0
        self.head = None
        self.tail = None

    def pop(self):
        """Removes the first item"""
        if self.length == 0:
            raise IndexError
        old_head = self.head
        self.head = old_head.next
        self.length -= 1
        if self.length == 0:
            self.tail = None
        return old_head
    
    def push(self, value):
        """Adds an item to the beginning"""
        
        old_head = self.head
        new_head = Node(value)
        new_head.next = old_head
        self.head = new_head
        self.length += 1
        if self.length == 1:
            self.tail = new_head
        return self
    
    
    def __str__(self):
        node = self.head
        node_list = []
        while node:
            node_list.append(node.value)
            node = node.next
        return '[' + ', '.join(map(str, node_list)) + ']'

        
        

sl = Stack()

sl.push(10)
sl.push(11) 
sl.push(12)
sl.push(13)
sl.push(14)
sl.push(15)
print(sl)
print(sl.pop())
print(sl.pop())
print(sl.pop())
print(sl.pop())

```

```python
class Node:
    
    def __init__(self, value):
        self.value = value
        self.next = None
    
    def __str__(self):
        return f"value: {self.value}"
    
    def __repr__(self):
        return f"Node( {self.value})"


class Queue:
    
    def __init__(self):
        self.length = 0
        self.head = None
        self.tail = None
        
    def enqueue(self, value):
        """Adds an item to the end"""
        
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1
        return self
    

    
    def dequeue(self):
        """Removes the first item"""
        if self.length == 0:
            raise IndexError
        old_head = self.head
        self.head = old_head.next
        self.length -= 1
        if self.length == 0:
            self.tail = None
        return old_head
    
    def __str__(self):
        node = self.head
        node_list = []
        while node:
            node_list.append(node.value)
            node = node.next
        return '[' + ', '.join(map(str, node_list)) + ']'

        
        

sl = Queue()

print(sl.enqueue(10))
print(sl.enqueue(11))
print(sl.enqueue(12))
print(sl.enqueue(13))
print(sl.dequeue())
print(sl)
print(sl.dequeue())
print(sl)
print(sl.dequeue())
print(sl)
print(sl.dequeue())
print(sl)
```
